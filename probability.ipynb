```python
# # Discrete Uniform Distribution

# Probability Mass Functions: Assgins probabilities to variables
# Most Basic Probability Model: Discrete Uniform Distrubution
# - All possible values of X must be consecutive integers (eg. 1-6 with dice)
# - All possible values must have an equal probability of occuring

# DUD Probability Mass: Example: single dice roll or single coin flip
# 1 / (b-a)+1  [ b: highest value , a: lowest value]
# eg. Dice roll: 1 / (6-1)+1 = 1/6
def dud_probability_mass(min_val,max_val):
    return 1 / ((max_val,min_val)+1)
# DUD Variance: How far a set of numbers is spread out:
def dud_variance(min_val,max_val):
    return ((max_val-min_val+2) * (max_val-min_val)) / 12

# Relative Frequency Histogram:
# Shows how often values are expected for each event (x axis:outcomes; y-axis:probability)
# Cumulative Distribution Chart:
# Shows cumulative probability associated with adding additional possible outcomes (x,y: outcomes, on-chart: cumulative probability, up to 1)
```

```python
# Expected Value:
# The long term expected average. 
# Found by summing multiples of x by probability of x.
roll_probabilites_two_dice = [
    [2 ,1/36], # Only one way (1 & 1)
    [3 ,2/36], # Two ways (1&2, 2&1)
    [4, 3/36], # Three ways (1&3, 3&1, 2&2)
    [5 ,4/36],
    [6 ,5/36],
    [7, 6/36],
    [8 ,5/36],
    [9 ,4/36],
    [10, 3/36],
    [11 ,2/36],
    [12, 1/36], # Only one way (6 & 6)
]  

def expected_value(outcome_and_probabilities):
    return sum([i[0]*i[1] for i in outcome_and_probabilities])

expected_value_2dice = expected_value(roll_probabilites_two_dice)
print(expected_value_2dice)
# Output: 6.999, which makes sense as 7 is the most likely outcome


# Variance with multiple variables:
# The expected variability in the results:
# V(x) = SUMALLX ( p(x) * (x-EV)**2)
def expected_variability(outcome_and_probabilities):
    return sum( [i[1] * ((i[0]-expected_value(outcome_and_probabilities))**2) for i in outcome_and_probabilities] )
# Variability:
print(expected_variability(roll_probabilites_two_dice))

# Standard Deviation
# Measure of the amount of variation
# Low SD indicates close to the mean, high SD indicates spread over large range
def std_dev(variance):
    return math.sqrt(variance)
# SD
print(std_dev(expected_variability(roll_probabilites_two_dice)))
```

```python
# # Binomial Probability
# Focuses on whether something will or will not occur
# Required conditions:
# multiple fixed trials with pass or fail
# each trial doesn't affect others
# trial probabilities are equal

# Eg. Rolling 1 die 100 times, binomial distribution of rolling 6 is possible
# (p of rolling die until five 6s are acheived is not a binomial dist problem though as no fixed trials)
# n: n_trials
# r: successes
# nCr: combinations
# p: probability of each value
# p(x=r) = nCr * (p**r) * ( (1-p)**(n-r) )
# p(x=r) =  nCr                 [binomial coefficient] * 
#           (p**r)              [p of success] * 
#           ( (1-p)**(n-r) )    [p of failure]
# eg. Probability of exactly four 6s in ten rolls
n = 10 # rolls/trials
r = 4 # successes
p = .167 # 1/6 
nCr = math.factorial(n) / (math.factorial(r) * math.factorial(n-r))
def binomial_probability(n,r,p,nCr):
    return nCr * (p**r) * ( (1-p)**(n-r) )
result = binomial_probability(n,r,p,nCr)
print(result)
```

```python
# # Poisson Distribution
# Focusses on probability that specific number of events will occur 
# over time, distance, area etc.
# Eg. number of visits to a website per day
# Required conditions:
# Count occurences over the unit of measurement (time,distance..)
# Mean occurence must be same over similar measurement
# Count over an interval can't affect other intervals
# Intervals can't overlap

# Eg. can grow 6 lettuce in 0.3sq meters
# p of growing 8?
# p(x) = ( (lda**x) * (eulers_num**-lda) ) / factorial(x)
eulers_number = 2.718 # this is a constant, like pi
def p_poisson(lda,x):
    return ( (lda**x) * (2.718**-lda) ) / math.factorial(x)
x = 8 # target is to grow 8
lda = 6 # lambda is 6, the amount that is usually grown in that area (.3sqm)
p_of_growing_8_when_6_is_average = p_poisson(lda,x)
p_of_growing_8_when_6_is_average
```

```python
# # Geometric Probability
# Used to find probability of success depending on number of trials
# Eg. how many trials required to roll a 6
# Required conditions:
# Independent trials
# Outcomes are either success or failure
# Constant probability of success
# n of trials need not be fixed
# --
# p(s=n) = p * ((1-p)**(n-1))
# p of success, multiplied by p of failure to the power of trials-1

# Eg. successful 6 in 3 dice rolls
p_success = 0.167
n = 3 # 3 rolls

# Probability of success on 3rd roll:
def p_geometric(p,n):
    return p * ((1-p)**(n-1))
p_on_3rd = p_geometric(p_success,n)
print(p_on_3rd)
# Probability of success on 3rd or 2nd or 1st roll (s<=n):
p_on_3rd_or_less = sum ([ p_geometric(p_success,n) for p_success,n in zip([0.167,0.167,0.167],[3,2,1])] )
print(p_on_3rd_or_less)
# Probability of more than 3 dice rolls:
print(f'Probability of success only after more than three dice rolls is 1-p of success on third: {1-p_geometric(p_success,n)}')

# Success is expected in approximately:
print(f'Simple expected formula, 1 divided by p of success. Success is expected after {1/p_success} trials')

print(f'Variance: {(1-p_success)/(p_success**2)}')
print(f'SD: {math.sqrt((1-p_success)/(p_success**2))}')
```

```python
# # Negative Binomial Probability
# While Geometric Model focuses on number of trials required for success
# Negative Binomial Model finds number of trials until N_th success
# Required conditions:
# Independent random trials (success or failure outcome in each)
# Each Trial has equal probability of success
# Track number of trials up to target success

# Binomial Model: When wanting to get probability of success within fixed number of trials
# Negative Binomial Model: When wanting to get probability of fixed number of successes in fixed number of trials
# Geometric Model: When wanting to find number of trials required for 1 single success

# Negative Binomial:
# p: p success
# k: target successes
# n: n of trials
# P(x=n) = c_with_repetition(k,n) * (p**k) * ( (1-p)**(n-k) )
# Eg. probability of rolling three 6s in ten rolls
n = 10 # 10 trials
k = 3 # want to have 3 success
p = .167 # 1/6
def negative_binomial(k,n,p):
    return c_with_repetition(k,n) * (p**k) * ( (1-p)**(n-k) )
outcome = negative_binomial(k,n,p)
print(f'Probability of 3 successful rolls of a certain dice number, within 10 trials is: {round(outcome*100,2)}%')
```

```python
# # Hypergeometric distribution
# Used for samples without replacement. Find probability a specific number of items fit a characteristic w/o replacement
# Binomial and Hypergeometric differ only in replacement
# Required Conditions:
# - Sample without replacement
# - Population has equal chance of being sampled
# - Population in 2 groups only (True/False or With/Without)
# - Total population known as well as people with the characteristic
# - X represents total people of interest
# Hypergeometric distribution:
# M = successes in population
# N = total population
# x = successes in sample
# n = total sample (from population)
# Formula:
# (combos_without_repetition(x,M) * combos_without_repetition(n-x,N-M)) / combos_without_repetition(n,N)
# Eg. Probability of getting 2 black cards out of 5 drawn from a deck - without replacement
N = 52 # total pop/cards
n = 5 # total sample
M = 26 # half of the cards in population/deck are black (52/2)
x = 2 # target is two black cards
def hypergeometric(population,successes_in_pop,sample,successes_in_sample):
    return (c_without_repetition(successes_in_sample,successes_in_pop) * c_without_repetition(sample-successes_in_sample,population-successes_in_pop)) / c_without_repetition(sample,population)
outcome = hypergeometric(N,M,n,x)
print(f'Probability of getting two black cards within 5 randomly drawn cards from a set of 52 is {round(outcome*100,1)}%')

print(f'Number of successes expected in the sample (expected value): {n*(M/N)}. This makes sense as half of the pack are black cards, so it would be expected that from 5 drawn cards, half would be black')
print(f'Average deviation from the mean over time (variance): {round(((n*M)/((N**2)*(N-1)))*(N-M)*(N-n),3)}')
print(f'Standard Deviation: {math.sqrt(round(((n*M)/((N**2)*(N-1)))*(N-M)*(N-n),3))}. So around 68% of the time we can expect 1.5 to 3.5 black cards')

```

```python
# # Continuous Probability
#
# Continuous Random Variable: Variables with infinite number of possible values
# Property Density Function: Provides density of concentration of probability at any point (x)
# Continuous Uniform Distribution: A continous random variable used when all values of the probability density function are the same or unknown
#
# CUD probabilities of success depending on all possible values between A-B, Used to find probability between an interval rather than a point.
# Ie. what is the probability that x is a value between 5 and 6
# Domain of X: Represents the interval that represents all possible values of x
# Density Function of x: Probability, is area under the curve.
# Conditions:
# All values of f(x) must be greater than or equal to 0
# Total area under the curve for all possible value of x should equal 1
# Formula:
# (interval difference) * (domain difference)
# (interval_max_val - interval_min_val) * (1/(domain_max_val - domain_min_val))
# Eg. Continuous Probability of interval(2,4) from a total domain (0,4):
interval = [2,4]
domain_of_x = [0,4]
def continuous_probability(interval,domain):
    return (interval[1]-interval[0]) * (1/(domain[1]-domain[0]))
outcome = continuous_probability(interval,domain_of_x)
print(f'Probability of x being in (2,4) when the domain of x stretches from 0 to 4 is: {round(outcome*100,3)}%. This is logical as half of the 4 units in 0-4 lie in 2-4')

```

```python
# # Exponential Distribution
# Exponential Function: A function of the form: f(x) = ab**x
# Exponential distribution: Continuous Distribution whose Probability Density has the shape of an Exponential Function
# Often used to model the time elapsed between events
# Density Function f(x) = lda(e)**(-lda(x)), where x>=0 and lda is constant
# Lda is the parameter of the exponential distribution. A large lambda is going to cause the slope of the curve to drop quick drop to zero, and vice versa.
# Probability of Exponential Distribution: Area under the curve
#
# Eg. Probability of getting home in 15 mins (P(X<0.25 hours))
# Formula for 'LESS than':
# P(X<x) = 1-e ** (-lda(x))
mean = 0.5 # 30 mins
lda = 1/mean
x = 0.25 # 15 mins
e = 2.718 # the exponential constant: common mathematical constant
def exponential_lessthan(x,lda):
    return 1 - (math.e**((-lda)*x))
outcome = exponential_lessthan(x,lda)
print(f'Exponential probability of getting home in less than 15 mins (less than half the usual amount of time): {round(outcome*100,2)}%')
# Probability of getting home in more than 15 mins
# Formula for 'MORE than':
# P(X>x) = e**((-lda)*x)
def exponential_morethan(x,lda):
    return math.e**((-lda)*x)
outcome = exponential_morethan(x,lda)
print(f'Exponential probability of getting home in more than 15 mins (half the usual amount of time or more): {round(outcome*100,2)}%')
# Probability of getting home in more than 15 mins but less than 40 mins
# Formula for 'BETWEEN':
# P(a<x<b) = e**((-lda)*a) - e**((-lda)*b)
a = 0.25 # 15 mins, lower boundary
b = 0.66 # 40 mins, upper boundary
def exponential_between(x,lda):
    return math.e**((-lda)*a) - math.e**((-lda)*b)
outcome = exponential_between(x,lda)
print(f'Exponential probability of getting home in between 15 and 40 mins (either side of the usual amount of time): {round(outcome*100,2)}%')
# Expected Value
print(f'Expected Value: {1/lda}. As expected this is the usual amount of time taken.')
print(f'Variance: {1/(lda**2)}. Average deviation from mean over time')
```
